```python
import matplotlib.pyplot as plt
import pandas as pd

def main():
    
    """ here I am going to discuss the average number of passengers travelled from 2018 to 2022 in Great Britain"""
   
    
    places=['avanti coast', 'midland railway', 'western railway']

# here avanti railway describes its position 
    x1 = [10, 20, 30, 40]
    y1 = [2018, 2019, 2020, 2021]
    plt.plot(x1, y1, label = 'avanti coast', marker='o')

# midland railway shows inclination
    x2 = [10, 40]
    y2 = [2020, 2022]
    plt.plot(x2, y2, label = 'midland railway', marker='*')

# here western railway best describes the travelling of passengers.
    x3 = [10, 50]
    y3 = [2020, 2019]
    plt.plot(x3, y3, label = 'western railway', marker='.')

# x labels shows the number of passengers travelling in billions
    plt.xlabel('passengers (billion)')

# here in y label I put years
    plt.ylabel('years')

# title describes the travelling of passengers in Britain
    plt.title('passengers kilometers by operators, Great Britain')

# legends helps to understand the different railways 
    plt.legend()

# to visualize the multiple line graph
    plt.show()
    
main()
```

```python
# showing CCTV survey Globally

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.getcwd()

def function1():
    
    """In this bar chart I want to show the CCTV survey in 2016, and data might be collected
        from four different countries globally."""
     
# place of file defined     
    df=pd.read_csv("watch.csv")
    
# defining Country list    
    Country = df['Country']
    
# here NS stands for Number Surveyed
    NS = df['NumberSurveyed']
    
# here HC stands for HassCCTV
    HC = df['HasCCTV']
    
# here IN stands for InUnloading  
    IN = df['InUnloading']
    
# here IL stands for InLairage 
    IL = df['InLairage']
    
# here IR stands for InRaceAndRestraint 
    IR = df['InRaceAndRestraint']
    
# here IS stands for InStun 
    IS = df['InStun']
    
# here IB stands for InBleed     
    IB = df['InBleed']
    
# let me describe bar width   
    width=0.1

# figure size    
    plt.figure(figsize=(12,7))
    
    xpos=np.arange(len(Country))
    
# setting all elements in one variable for providing proper legends    
    y=['NS=NumberSurveyed','HC=HasCCTV','IN=InUnloading','IL=InLairage','IR=InRaceAndRestraint','IS=InStun','IB=InBleed']


# describing bar position, width, color and label
    plt.bar(xpos,NS,color='aqua',width=width,label=NS)
    
    plt.bar(xpos+0.10,HC,color='steelblue',width=width,label=HC)
    
    plt.bar(xpos+0.2,IN,color='thistle',width=width,label=IN)
    
    plt.bar(xpos+0.3,IL,color='salmon',width=width,label=IL)
    
    plt.bar(xpos+0.4,IR,color='palegreen',width=width,label=IS)
    
    plt.bar(xpos+0.5,IS,color='violet',width=width,label=IB)
    
    plt.bar(xpos+0.6,IB,color='lightseagreen',width=width,label=IB)
    
    plt.xticks(xpos+0.3,('England','Turkey','Wales','America'))
    
# providing x labels
    plt.xlabel('Global Survey')
    
# providing y labels    
    plt.ylabel('Entries Recorded')
    
# title describes the bar graph related to CCTV survey
    plt.title('CCTV Survey 2016')
    
    plt.legend(y)
    
# to show figure
    plt.show()
    
function1()


```
